```python
# this allows plots to appear directly in the notebook
%config InlineBackend.figure_format='retina'
%matplotlib inline

# imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('ads.csv', index_col=0)
df.head()
```

```python
fig, axs = plt.subplots(1, 3, sharey=True);
df.plot(kind='scatter', x='tv', y='sales', ax=axs[0], figsize=(16, 8));
df.plot(kind='scatter', x='radio', y='sales', ax=axs[1]);
df.plot(kind='scatter', x='newspaper', y='sales', ax=axs[2]);
```

```python
# this is the standard import if you're using "formula notation" (similar to R)
import statsmodels.formula.api as smf

# create a fitted model in one line
lm = smf.ols(formula='sales ~ tv', data=df).fit()

# print the coefficients
lm.params
```

```python
7.032594 + 0.047537*100
```

```python
x_pred = pd.DataFrame({'tv': [100]})
lm.predict(x_pred)
```

```python
# create a fitted model in one line
lm2 = smf.ols(formula='sales ~ radio', data=df).fit()

# print the coefficients
lm2.params
```

```python
# create a fitted model in one line
lm3 = smf.ols(formula='sales ~ newspaper', data=df).fit()

# print the coefficients
lm3.params
```

```python
sns.lmplot(x="tv", y="sales",data=df, height=7, aspect=1.5)
```

```python
lm.rsquared
```

```python
# create a fitted model in one line
lm4 = smf.ols(formula='sales ~ tv + radio + newspaper', data=df).fit()

# print the coefficients
lm4.params
```

```python
lm4.rsquared
```

```python
from sklearn.cross_validation import cross_val_score

feature_cols = ['tv', 'radio', 'newspaper']
X = df[feature_cols]
y = df['sales']

scores = cross_val_score(lm, X, y, cv=10, scoring='neg_mean_squared_error')
print(scores.mean())
```

```python
# create X and y
feature_cols = ['tv', 'radio', 'newspaper']
X = df[feature_cols]
y = df['sales']

# follow the usual sklearn pattern: import, instantiate, fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X, y)

# print intercept and coefficients
print ("intercept", lm.intercept_)
print ("coefficients:", lm.coef_)
```

```python
# create X and y
feature_cols = ['tv', 'radio', 'newspaper', 'is_large']
X = df[feature_cols]
y = df['sales']

# instantiate, fit
lm = LinearRegression()
lm.fit(X, y)

# print coefficients
list(zip(feature_cols, lm.coef_))
```

```python
scores = cross_val_score(lm, X, y, cv=10, scoring='neg_mean_squared_error')
print(scores.mean())
```
